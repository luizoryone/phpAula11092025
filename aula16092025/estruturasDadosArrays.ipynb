## Resenha Geral sobre Arrays em JavaScript

Arrays são a espinha dorsal de muitas estruturas de dados em programação. Em JavaScript, eles se destacam pela sua **flexibilidade**, permitindo armazenar uma coleção **ordenada** de valores, que podem ser de **diferentes tipos de dados**. Essa capacidade torna os arrays ferramentas poderosas para organizar e manipular informações de maneira eficiente. Pense neles como recipientes inteligentes, onde cada item tem sua posição definida, facilitando o acesso e a modificação. Embora JavaScript permita a mistura de tipos, a **boa prática** sugere manter a uniformidade para simplificar o código e evitar erros inesperados.

-----

## Glossário de Termos Relacionados a Arrays

  * **Array:** Uma estrutura de dados que armazena uma coleção ordenada de elementos.
  * **Elemento:** Um valor individual dentro de um array.
  * **Índice:** A posição numérica de um elemento em um array, começando em 0.
  * **Declaração:** O ato de criar uma variável para armazenar um array.
  * **Inicialização:** O ato de preencher um array com valores iniciais.
  * **Propriedade `length`:** Retorna o número de elementos em um array.
  * **Método:** Uma função associada a um objeto (neste caso, arrays) que realiza uma operação.
  * **Iteração:** O processo de percorrer todos os elementos de um array.
  * **Cópia Rasa (Shallow Copy):** Uma cópia de um array onde os elementos que são objetos ou outros arrays são referenciados, não duplicados.

## Melhores Práticas e O Que Não Fazer com Arrays

### Melhores Práticas

  * **Uniformidade de Tipos:** Sempre que possível, mantenha um único tipo de dado em um array. Isso aumenta a legibilidade e reduz a probabilidade de erros.
      * **Exemplo:** `let numeros = [1, 2, 3];` é melhor que `let misto = [1, "dois", true];` se a intenção for trabalhar com números.
  * **Utilize `const` para Arrays que Não Serão Reatribuídos:** Se você declara um array com `let` e nunca o reatribui a um novo array, considere usar `const`. Isso não impede a modificação dos elementos internos, mas garante que a variável sempre aponte para o mesmo array.
      * **Exemplo:**
        ```javascript
        const cores = ["vermelho", "azul"];
        cores.push("verde"); // Válido
        // cores = ["amarelo"]; // Inválido com const
        ```
  * **Preferir Métodos Modernos:** Utilize métodos como `forEach`, `map`, `filter`, `reduce` para iterações e transformações, pois geralmente resultam em código mais limpo e expressivo.
  * **Evitar Índices Negativos:** Embora alguns métodos aceitem índices negativos (como `slice`), evite usá-los diretamente ao acessar elementos para manter a clareza, a menos que seja intencional para funcionalidades específicas.
  * **Compreender a Imutabilidade:** Ao usar métodos como `slice` e `filter`, você obtém novos arrays, preservando o original. Ao usar métodos como `splice` e `push`, o array original é modificado. Tenha clareza sobre qual comportamento você precisa.

### O Que Não Fazer

  * **Não Acessar Índices Inexistentes Constantemente:** Acessar um índice que está fora dos limites do array resulta em `undefined`. Embora isso possa ser usado intencionalmente, o uso frequente pode indicar um erro lógico.
  * **Não Modificar Arrays Grandes em Loops Desnecessariamente:** Operações que alteram o tamanho do array dentro de um loop `for` tradicional (como `push` ou `splice` que aumentam o array) podem levar a comportamentos inesperados ou a loops infinitos se não forem controladas cuidadosamente. Prefira criar um novo array com os resultados.
  * **Não Confundir `slice()` com `splice()`:** Lembre-se que `slice()` retorna uma cópia sem modificar o original, enquanto `splice()` modifica o array original.
  * **Não Usar `eval()` com Índices ou Conteúdo de Arrays:** Evite usar `eval()` para acessar ou manipular elementos de arrays, pois isso representa um risco de segurança e geralmente é desnecessário.

## Melhores Abordagens e Uso Profissional

Arrays são onipresentes no desenvolvimento profissional. Suas aplicações são vastas:

1.  **Armazenamento de Dados:** Desde listas de usuários, produtos, pedidos, até configurações de um aplicativo.
2.  **Manipulação de Sequências:** Em processamento de texto, análise de dados, e algoritmos que operam sobre sequências.
3.  **Modelagem de Estruturas:** Representar tabelas, grades, filas, pilhas, etc.
4.  **Retorno de Múltiplos Valores:** Funções podem retornar arrays para fornecer diversos resultados.
5.  **Gerenciamento de Estados:** Em frameworks de frontend como React ou Vue, arrays são frequentemente usados para gerenciar listas de componentes ou dados exibidos na interface.

### Exemplos de Uso Profissional

#### 1\. Processamento de Vendas Diárias

Imagine um sistema de e-commerce. Você pode ter um array para armazenar o valor de cada venda realizada em um dia:

```javascript
let vendasDia = [150.75, 230.00, 85.50, 410.20, 199.99];

// Calcular o total de vendas
let totalVendas = 0;
for (let i = 0; i < vendasDia.length; i++) {
    totalVendas += vendasDia[i];
}
console.log(`Total de vendas do dia: R$ ${totalVendas.toFixed(2)}`);

// Adicionar uma nova venda
let novaVenda = 55.00;
vendasDia.push(novaVenda);
console.log(`Vendas após nova adição: ${vendasDia}`);

// Verificar se houve alguma venda acima de R$ 400
if (vendasDia.includes(vendasDia.find(venda => venda > 400))) {
    console.log("Houve pelo menos uma venda acima de R$ 400!");
}
```

#### 2\. Gerenciamento de Tarefas (To-Do List)

Um array pode representar a lista de tarefas pendentes:

```javascript
let tarefasPendentes = ["Comprar pão", "Estudar JavaScript", "Ligar para o cliente"];

// Adicionar nova tarefa
tarefasPendentes.unshift("Agendar reunião"); // Adiciona no início por ser mais urgente

// Marcar tarefa como concluída (remover)
let tarefaConcluida = tarefasPendentes.shift(); // Remove a primeira tarefa
console.log(`Tarefa concluída: ${tarefaConcluida}`);
console.log(`Tarefas restantes: ${tarefasPendentes}`);

// Exibir todas as tarefas
console.log("Lista de Tarefas:");
tarefasPendentes.forEach((tarefa, index) => {
    console.log(`${index + 1}. ${tarefa}`);
});
```

#### 3\. Configurações de Permissões de Usuário

Em um sistema de permissões, um array pode listar as ações permitidas para um determinado papel:

```javascript
let permissoesAdmin = ["criar_usuario", "editar_usuario", "excluir_usuario", "ver_logs"];
let permissoesEditor = ["criar_post", "editar_post", "ver_posts"];

function verificarPermissao(permissoes, acao) {
    return permissoes.includes(acao);
}

console.log(`Admin pode criar usuário? ${verificarPermissao(permissoesAdmin, "criar_usuario")}`); // true
console.log(`Editor pode excluir usuário? ${verificarPermissao(permissoesEditor, "excluir_usuario")}`); // false
```

### Exercício 2: Gerenciamento de Notas de Alunos

**Enunciado:** Um array `notas` contém as seguintes notas de um aluno: `[7.5, 8.0, 6.5, 9.0]`. Calcule a média das notas. Se a média for menor que 7.0, adicione a nota `5.0` ao final do array para uma recuperação e recalcule a média.

**Resposta:**

```javascript
let notas = [7.5, 8.0, 6.5, 9.0];

function calcularMedia(arrNotas) {
    let soma = 0;
    for (let i = 0; i < arrNotas.length; i++) {
        soma += arrNotas[i];
    }
    return soma / arrNotas.length;
}

let mediaAtual = calcularMedia(notas);
console.log(`Média inicial: ${mediaAtual.toFixed(2)}`);

if (mediaAtual < 7.0) {
    console.log("Média abaixo de 7.0. Adicionando nota para recuperação.");
    notas.push(5.0);
    mediaAtual = calcularMedia(notas);
    console.log(`Notas após recuperação: ${notas}`);
    console.log(`Nova média: ${mediaAtual.toFixed(2)}`);
} else {
    console.log("Média satisfatória. Nenhuma recuperação necessária.");
}
```

### Resenha: `push()` e `pop()` - Gerenciamento de Pilha (Stack)

**Conceito:** Os métodos `push()` e `pop()` são ideais para implementar uma estrutura de dados tipo **pilha (stack)**, que segue o princípio LIFO (Last-In, First-Out – o último a entrar é o primeiro a sair).

**Aplicação:** Imagine um histórico de navegação em um navegador. Cada página visitada é `push`ada na pilha. Quando o usuário clica em "Voltar", a página atual é `pop`ada, revelando a anterior.

**Exemplo Detalhado:**

```javascript
let historicoNavegacao = [];

// Visitando páginas
historicoNavegacao.push("Pagina Inicial"); // ['Pagina Inicial']
historicoNavegacao.push("Pagina de Produtos"); // ['Pagina Inicial', 'Pagina de Produtos']
historicoNavegacao.push("Pagina de Detalhes do Produto X"); // ['Pagina Inicial', 'Pagina de Produtos', 'Pagina de Detalhes do Produto X']

console.log("Histórico atual:", historicoNavegacao);

// Usuário clica em "Voltar"
let paginaAnterior = historicoNavegacao.pop(); // Remove e retorna 'Pagina de Detalhes do Produto X'
console.log(`Voltando de: ${paginaAnterior}`);
console.log("Histórico após voltar:", historicoNavegacao); // ['Pagina Inicial', 'Pagina de Produtos']

paginaAnterior = historicoNavegacao.pop(); // Remove e retorna 'Pagina de Produtos'
console.log(`Voltando de: ${paginaAnterior}`);
console.log("Histórico após voltar novamente:", historicoNavegacao); // ['Pagina Inicial']

// Se o usuário tentar voltar mais uma vez
let ultimaPagina = historicoNavegacao.pop(); // Remove e retorna 'Pagina Inicial'
console.log(`Voltando de: ${ultimaPagina}`);
console.log("Histórico após última volta:", historicoNavegacao); // []

// Tentativa de voltar de um histórico vazio
let nadaMais = historicoNavegacao.pop(); // Retorna undefined
console.log(`Tentativa de voltar de histórico vazio: ${nadaMais}`); // undefined
console.log("Histórico final:", historicoNavegacao); // []
```

**Esgotando os Meios de Explicação:** `push()` adiciona um ou múltiplos elementos ao final, retornando o novo comprimento. `pop()` remove apenas o último elemento e o retorna. Ambos modificam o array original e são essenciais para gerenciar coleções onde a ordem de adição/remoção importa de forma LIFO. São extremamente eficientes para suas tarefas.

### Resenha: `shift()` e `unshift()` - Gerenciamento de Filas (Queues)

**Conceito:** `shift()` e `unshift()` operam nas extremidades opostas de um array em relação a `pop()` e `push()`. Eles são ideais para implementar uma estrutura de dados tipo **fila (queue)**, que segue o princípio FIFO (First-In, First-Out – o primeiro a entrar é o primeiro a sair).

**Aplicação:** Filas de atendimento em call centers, processamento de tarefas em ordem de chegada, ou a ordem em que os jogadores entram em uma partida.

**Exemplo Detalhado:**

```javascript
let filaAtendimento = [];

// Novos clientes chegam
filaAtendimento.unshift("Cliente A"); // ['Cliente A']
filaAtendimento.unshift("Cliente B"); // ['Cliente B', 'Cliente A'] - Novo cliente é prioridade máxima (início da fila)
filaAtendimento.unshift("Cliente C"); // ['Cliente C', 'Cliente B', 'Cliente A']

console.log("Fila de atendimento (ordem de chegada):", filaAtendimento); // Note que a ordem de UNSHIFT é invertida

// Atendendo o cliente que chegou primeiro (será o último do array criado com unshift)
// Para simular FIFO corretamente com unshift/shift, o primeiro a chegar deve ser o último adicionado com unshift
// Vamos refazer com a lógica correta de FIFO:

let filaAtendimentoFIFO = [];

// Clientes chegam e são adicionados ao FINAL da fila (como em uma fila real)
filaAtendimentoFIFO.push("Cliente 1"); // ['Cliente 1']
filaAtendimentoFIFO.push("Cliente 2"); // ['Cliente 1', 'Cliente 2']
filaAtendimentoFIFO.push("Cliente 3"); // ['Cliente 1', 'Cliente 2', 'Cliente 3']

console.log("Fila de atendimento real (FIFO):", filaAtendimentoFIFO);

// Atendendo o cliente no INÍCIO da fila
let clienteSendoAtendido = filaAtendimentoFIFO.shift(); // Remove e retorna 'Cliente 1'
console.log(`Atendendo: ${clienteSendoAtendido}`);
console.log("Fila após atendimento:", filaAtendimentoFIFO); // ['Cliente 2', 'Cliente 3']

clienteSendoAtendido = filaAtendimentoFIFO.shift(); // Remove e retorna 'Cliente 2'
console.log(`Atendendo: ${clienteSendoAtendido}`);
console.log("Fila após segundo atendimento:", filaAtendimentoFIFO); // ['Cliente 3']

// Um novo cliente chega e é colocado no final
filaAtendimentoFIFO.push("Cliente 4");
console.log("Fila após chegada de novo cliente:", filaAtendimentoFIFO); // ['Cliente 3', 'Cliente 4']

// Agora, usando unshift para adicionar uma tarefa de ALTA PRIORIDADE no INÍCIO da fila de tarefas
let filaDeTarefas = ["Enviar Relatório Mensal", "Responder E-mails"];
console.log("Tarefas antes da prioridade:", filaDeTarefas);

filaDeTarefas.unshift("Resolver Bug Crítico"); // Adiciona no início
console.log("Tarefas com prioridade:", filaDeTarefas); // ['Resolver Bug Crítico', 'Enviar Relatório Mensal', 'Responder E-mails']

let proximaTarefa = filaDeTarefas.shift(); // Remove 'Resolver Bug Crítico'
console.log(`Executando tarefa de alta prioridade: ${proximaTarefa}`);
console.log("Tarefas restantes:", filaDeTarefas); // ['Enviar Relatório Mensal', 'Responder E-mails']
```

**Esgotando os Meios de Explicação:** `shift()` remove o primeiro elemento e o retorna, deslocando todos os outros para a esquerda. `unshift()` adiciona um ou múltiplos elementos no início, deslocando os existentes para a direita, e retorna o novo comprimento. Ambos são cruciais para operações FIFO e para gerenciar elementos com prioridade no início de uma lista.

### Resenha: `splice()` - O Poder da Manipulação Direta

**Conceito:** `splice()` é o método mais poderoso para modificar um array diretamente, permitindo **remover**, **adicionar** ou **substituir** elementos em qualquer posição. Ele altera o array original e retorna os elementos que foram removidos.

**Aplicação:** Edição de listas em interfaces de usuário (remover um item, adicionar um novo em uma posição específica), reordenamento dinâmico de dados, ou até mesmo para simular a inserção/remoção em qualquer ponto de uma sequência.

**Exemplo Detalhado (Abrangente):**

```javascript
let jogadores = ["Alice", "Bob", "Charlie", "David", "Eve"];
console.log("Jogadores iniciais:", jogadores);

// 1. Remover um jogador (Bob)
// splice(indice_inicio, quantidade_remover)
let removed = jogadores.splice(1, 1); // Remove 1 elemento a partir do índice 1
console.log("Jogador removido:", removed); // ['Bob']
console.log("Jogadores após remoção:", jogadores); // ['Alice', 'Charlie', 'David', 'Eve']

// 2. Adicionar dois novos jogadores (Frank, Grace) na posição onde Bob estava
// splice(indice_inicio, quantidade_remover, item1, item2, ...)
jogadores.splice(1, 0, "Frank", "Grace"); // Remove 0 elementos, adiciona 'Frank' e 'Grace' no índice 1
console.log("Jogadores após adicionar Frank e Grace:", jogadores); // ['Alice', 'Frank', 'Grace', 'Charlie', 'David', 'Eve']

// 3. Substituir Charlie por Heidi
// splice(indice_inicio, quantidade_remover, item_novo)
jogadores.splice(3, 1, "Heidi"); // Remove 1 elemento no índice 3 ('Charlie') e insere 'Heidi'
console.log("Jogadores após substituir Charlie por Heidi:", jogadores); // ['Alice', 'Frank', 'Grace', 'Heidi', 'David', 'Eve']

// 4. Remover os últimos dois jogadores (David, Eve) de uma vez
// Podemos usar índices negativos para contar do final
jogadores.splice(-2, 2); // Remove 2 elementos a partir dos últimos 2
console.log("Jogadores após remover os últimos dois:", jogadores); // ['Alice', 'Frank', 'Grace', 'Heidi']

// 5. Adicionar um jogador no início sem remover nada
jogadores.splice(0, 0, "Início");
console.log("Jogadores com 'Início' adicionado:", jogadores); // ['Início', 'Alice', 'Frank', 'Grace', 'Heidi']

// 6. Remover todos os elementos do array
jogadores.splice(0, jogadores.length);
console.log("Jogadores após remover todos:", jogadores); // []

// 7. Adicionar múltiplos elementos ao array vazio
jogadores.splice(0, 0, "A", "B", "C", "D");
console.log("Jogadores após adicionar vários em array vazio:", jogadores); // ['A', 'B', 'C', 'D']

// Exemplo avançado: Inserir e Remover na mesma operação de forma mais complexa
let numerosComplexos = [10, 20, 30, 40, 50];
// Remover 30 e 40 (índice 2, 3) e inserir 35 entre 20 e 50
// Queremos remover 2 elementos a partir do índice 2
// E queremos inserir 35 na posição do índice 2 (após o 20)
let removidosComplexos = numerosComplexos.splice(2, 2, 35); // Remove 40, 50 e insere 35 na posição 2
// Ops, a descrição do splice é: start, deleteCount, item1, item2...
// Se eu quero remover 30 e 40 (índices 2 e 3), então deleteCount é 2
// Se eu quero inserir 35 na posição onde 30 estava (índice 2), então o splice deve começar no índice 2
console.log("Array original antes de complexo:", [10, 20, 30, 40, 50]);
let arrayModificado = [10, 20, 30, 40, 50];
let removidos = arrayModificado.splice(2, 2, 35); // Remove 30 e 40, insere 35 na posição 2
console.log("Array modificado:", arrayModificado); // [ 10, 20, 35, 50 ]
console.log("Removidos:", removidos); // [ 30, 40 ]
```

**Esgotando os Meios de Explicação:** `splice()` é extremamente flexível. Seu poder reside na combinação dos três parâmetros: o **índice de início** determina onde a mudança ocorre; `deleteCount` define quantos elementos remover; e os itens subsequentes são os novos elementos a serem inseridos. Se `deleteCount` for `0`, apenas adiciona; se não houver itens adicionais, apenas remove. A capacidade de operar em qualquer índice e a flexibilidade em adicionar/remover o tornam uma ferramenta essencial para manipulações complexas de arrays.

### Resenha: `slice()` - Criando Cópias e Subconjuntos

**Conceito:** Ao contrário de `splice()`, o método `slice()` é **não destrutivo**. Ele retorna um **novo array** que é uma cópia rasa de uma porção do array original, sem alterar o array original.

**Aplicação:** Obter partes de um array para processamento independente, criar cópias de segurança de arrays antes de modificá-los, ou extrair dados para exibição em diferentes partes de uma interface.

**Exemplo Detalhado:**

```javascript
let componentesPC = ["CPU", "Placa Mãe", "Memória RAM", "SSD", "Placa de Vídeo", "Fonte"];
console.log("Componentes originais:", componentesPC);

// 1. Obter uma cópia exata do array original
let copiaCompleta = componentesPC.slice();
console.log("Cópia completa do array:", copiaCompleta);
console.log("Array original permanece inalterado:", componentesPC);

// 2. Obter os componentes de armazenamento (SSD) e fonte
// SSD está no índice 3, Fonte no índice 5. Queremos do 3 até o 6 (exclusive)
let armazenamento = componentesPC.slice(3, 6);
console.log("Componentes de armazenamento:", armazenamento);

// 3. Obter todos os componentes a partir da Placa Mãe (índice 1)
let doMeioParaFrente = componentesPC.slice(1);
console.log("Componentes a partir da Placa Mãe:", doMeioParaFrente);

// 4. Usando índices negativos para contar do fim
// Queremos obter os dois últimos componentes (Placa de Vídeo e Fonte)
// slice(-2) começa a partir do segundo elemento contando do final
let ultimosDois = componentesPC.slice(-2);
console.log("Últimos dois componentes:", ultimosDois);

// 5. Combinando índices positivos e negativos
// Queremos do índice 2 ("Memória RAM") até o penúltimo elemento
let parteDoMeio = componentesPC.slice(2, -1); // Do índice 2 até o elemento antes do último
console.log("Parte do meio (índice 2 até penúltimo):", parteDoMeio);

// 6. O que acontece se start for maior que end?
let resultadoInesperado = componentesPC.slice(4, 2); // start=4, end=2
console.log("Resultado de slice(4, 2):", resultadoInesperado); // Retorna um array vazio []

// 7. Cópia rasa e mutabilidade de objetos/arrays aninhados
let dadosAninhados = [
    { id: 1, nome: "Item A" },
    [10, 20],
    { id: 2, nome: "Item B" }
];
let copiaDadosAninhados = dadosAninhados.slice();

console.log("Cópia aninhada:", copiaDadosAninhados);

// Modificando um elemento que é um objeto na cópia
copiaDadosAninhados[0].nome = "Item A Modificado";
console.log("Cópia aninhada após modificar objeto:", copiaDadosAninhados);
console.log("Original após modificar objeto na cópia:", dadosAninhados); // O original TAMBÉM é modificado!

// Modificando um elemento que é um array na cópia
copiaDadosAninhados[1].push(30);
console.log("Cópia aninhada após modificar array:", copiaDadosAninhados);
console.log("Original após modificar array na cópia:", dadosAninhados); // O original TAMBÉM é modificado!

// Adicionando um novo objeto à cópia
copiaDadosAninhados.push({ id: 3, nome: "Item C" });
console.log("Cópia aninhada após adicionar novo objeto:", copiaDadosAninhados);
console.log("Original após adicionar novo objeto à cópia:", dadosAninhados); // O original NÃO é modificado nesse caso!
```

**Esgotando os Meios de Explicação:** `slice()` é a ferramenta preferida quando você precisa de uma **subseção** ou uma **cópia** de um array sem afetar o original. Os argumentos `start` e `end` definem o intervalo (lembrando que `end` é exclusivo). A ausência de argumentos copia o array inteiro. É fundamental entender que `slice()` faz uma **cópia rasa**: objetos ou outros arrays dentro do array copiado ainda referenciam os mesmos objetos do array original. Modificações nesses objetos/arrays aninhados na cópia afetarão o original, mas adicionar ou remover elementos no array copiado não afeta o original.

## Perguntas e Respostas Adicionais

**P1: Qual a diferença principal entre `for...of` e `forEach()` para iterar sobre arrays?**

**R1:**

  * `for...of`: Permite iterar diretamente sobre os **valores** dos elementos em estruturas iteráveis (incluindo arrays). Oferece mais controle, como a possibilidade de usar `break` para sair do loop ou `continue` para pular uma iteração.
  * `forEach()`: Executa uma função de callback para cada elemento do array. É mais conciso para operações simples, mas **não permite** sair do loop com `break` (você teria que usar `return` dentro do callback, que funcionaria como `continue` para aquela iteração específica) e não oferece acesso direto ao índice de forma tão fluida quanto um `for` tradicional (embora o callback de `forEach` receba o índice como segundo argumento).

**P2: É possível ter um array de arrays (matriz) em JavaScript? Como acessá-lo?**

**R2:** Sim, é perfeitamente possível e muito comum. Um array de arrays é chamado de **matriz bidimensional** ou simplesmente matriz. Você acessa os elementos usando múltiplos índices: o primeiro índice para selecionar o array interno, e o segundo para selecionar o elemento dentro desse array interno.

**Exemplo:**

```javascript
let matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
];

console.log(matriz[0][0]); // Saída: 1 (primeiro array, primeiro elemento)
console.log(matriz[1][2]); // Saída: 6 (segundo array, terceiro elemento)
console.log(matriz[2][1]); // Saída: 8 (terceiro array, segundo elemento)

// Iterando sobre uma matriz
for (let i = 0; i < matriz.length; i++) {
    for (let j = 0; j < matriz[i].length; j++) {
        console.log(`Elemento na posição [${i}][${j}]: ${matriz[i][j]}`);
    }
}
```

**P3: O que acontece se eu tentar adicionar um elemento a um array usando um índice muito maior que o tamanho atual?**

**R3:** O JavaScript **não lança um erro**. Ele cria um array "esparso" ou "disjunto". Os elementos entre o último elemento existente e o índice que você especificou serão preenchidos com `undefined`. A propriedade `length` do array será atualizada para refletir o maior índice mais um.

**Exemplo:**

```javascript
let meuArray = ["a", "b"];
console.log("Array inicial:", meuArray, "Length:", meuArray.length); // [ 'a', 'b' ] Length: 2

meuArray[5] = "f"; // Adiciona 'f' no índice 5
console.log("Array após adicionar índice 5:", meuArray, "Length:", meuArray.length); // [ 'a', 'b', <3 empty items>, 'f' ] Length: 6

console.log(meuArray[2]); // Saída: undefined
console.log(meuArray[3]); // Saída: undefined
console.log(meuArray[4]); // Saída: undefined
```

Embora o JavaScript permita isso, geralmente é considerada uma **má prática** e pode levar a confusão. Métodos como `push()` ou `splice()` são preferíveis para adicionar elementos de forma controlada.

**P4: Posso "desestruturar" um array em variáveis individuais?**

**R4:** Sim\! A **desestruturação de arrays** (Array Destructuring) é uma funcionalidade moderna do JavaScript (ES6+) que permite extrair valores de arrays e atribuí-los a variáveis de forma concisa.

**Exemplo:**

```javascript
let coordenadas = [10.5, 20.2];

// Desestruturação tradicional
let x = coordenadas[0];
let y = coordenadas[1];
console.log(`x: ${x}, y: ${y}`); // x: 10.5, y: 20.2

// Desestruturação com ES6
let [cordX, cordY] = coordenadas;
console.log(`cordX: ${cordX}, cordY: ${cordY}`); // cordX: 10.5, cordY: 20.2

// Ignorar elementos
let [primeiro, , terceiro] = ["Maçã", "Banana", "Laranja", "Uva"];
console.log(primeiro); // Maçã
console.log(terceiro); // Laranja

// Usar valor padrão
let [nomeUsuario, emailUsuario = "sem_email@exemplo.com"] = ["Carlos"];
console.log(nomeUsuario); // Carlos
console.log(emailUsuario); // sem_email@exemplo.com

let [nomeUsuario2, emailUsuario2 = "sem_email@exemplo.com"] = ["Ana", "ana@exemplo.com"];
console.log(nomeUsuario2); // Ana
console.log(emailUsuario2); // ana@exemplo.com
```

# 🧠 Métodos Avançados de Arrays em JavaScript

Estes métodos fazem parte da abordagem **funcional** da programação, altamente valorizada no desenvolvimento moderno por sua legibilidade, reusabilidade e poder expressivo.

---

## 🔁 `map()` – Transformação de Arrays

### ✅ Descrição:

Cria um novo array com os **valores transformados** a partir do array original, **sem modificar o original**.

### 📌 Sintaxe:

```js
let novoArray = arrayOriginal.map((elemento, indice, array) => {
    // retorna novo valor
});
```

### 🧪 Exemplo:

```js
let numeros = [1, 2, 3];
let dobrados = numeros.map(n => n * 2);
console.log(dobrados); // [2, 4, 6]
```

### ✅ Aplicações profissionais:

* Formatar dados para exibição
* Aplicar transformações em massa
* Preparar payloads para APIs

---

## 🔍 `filter()` – Filtragem de Arrays

### ✅ Descrição:

Retorna um novo array **apenas com os elementos que passarem em uma condição**, **sem modificar o array original**.

### 📌 Sintaxe:

```js
let filtrados = arrayOriginal.filter((elemento, indice, array) => {
    return condiçãoBooleana;
});
```

### 🧪 Exemplo:

```js
let numeros = [10, 25, 30, 5];
let maioresQue20 = numeros.filter(n => n > 20);
console.log(maioresQue20); // [25, 30]
```

### ✅ Aplicações profissionais:

* Filtrar resultados de pesquisa
* Separar dados válidos ou inválidos
* Criar sublistas baseadas em critérios dinâmicos

---

## 🔄 `reduce()` – Redução/Agrupamento

### ✅ Descrição:

Executa uma função "acumuladora" em cada elemento do array, resultando em **um único valor** de saída.

### 📌 Sintaxe:

```js
let resultado = arrayOriginal.reduce((acumulador, elementoAtual, indice, array) => {
    return novoAcumulador;
}, valorInicial);
```

### 🧪 Exemplo: Soma de números

```js
let numeros = [1, 2, 3, 4];
let soma = numeros.reduce((total, numero) => total + numero, 0);
console.log(soma); // 10
```

### 🧪 Exemplo: Contar palavras

```js
let palavras = ["maçã", "banana", "maçã", "laranja", "banana"];
let contagem = palavras.reduce((acc, fruta) => {
    acc[fruta] = (acc[fruta] || 0) + 1;
    return acc;
}, {});
console.log(contagem);
// { maçã: 2, banana: 2, laranja: 1 }
```

### ✅ Aplicações profissionais:

* Somatórios, médias, totais
* Construção de objetos agrupados
* Conversão de arrays em outros formatos

## ❓ PERGUNTAS & RESPOSTAS INÉDITAS SOBRE MAP, FILTER E REDUCE

### 📌 1. Posso usar `map()` para filtrar dados?

Não é recomendado. `map()` sempre retorna a mesma quantidade de elementos. Use `filter()` para isso.

---

### 📌 2. `reduce()` sempre precisa de `valorInicial`?

Não obrigatoriamente, mas **é recomendado**, especialmente se o array estiver vazio ou o tipo do acumulador for diferente do primeiro item.

### 📌 3. Como faço para aplicar `map()` e depois `filter()`?

Você pode encadear:

```js

```

### 📌 4. `reduce()` pode retornar objetos?

Sim. É uma das formas mais poderosas de usar `reduce()`:

```js
let lista = ["a", "b", "a"];
let freq = lista.reduce((acc, letra) => {
    acc[letra] = (acc[letra] || 0) + 1;
    return acc;
}, {});
```

```jsx
// src/components/Carrinho.js
import React, { useEffect, useState } from 'react';

// Simulação de "API" local
const produtosMock = [
  { id: 1, nome: "Notebook", preco: 3000, emEstoque: true },
  { id: 2, nome: "Mouse", preco: 150, emEstoque: true },
  { id: 3, nome: "Teclado", preco: 230, emEstoque: false },
  { id: 4, nome: "Monitor", preco: 1200, emEstoque: true },
  { id: 5, nome: "HD Externo", preco: 500, emEstoque: false },
];

const cupons = {
  DESCONTO10: 0.10,
  DESCONTO20: 0.20
};

const formatarPreco = valor =>
  `R$ ${valor.toFixed(2).replace('.', ',')}`;

const calcularFrete = total =>
  total >= 2000 ? 0 : 100;

function Carrinho() {
  const [produtos, setProdutos] = useState([]);
  const [cupom, setCupom] = useState('');
  const [resumo, setResumo] = useState(null);

  // Carregar do "localStorage" ou mock
  useEffect(() => {
    const cache = localStorage.getItem("carrinho");
    if (cache) {
      setProdutos(JSON.parse(cache));
    } else {
      setProdutos(produtosMock);
    }
  }, []);

    localStorage.setItem("carrinho", JSON.stringify(produtos));
  }, [produtos]);

  useEffect(() => {
    calcularResumo();
  }, [produtos, cupom]);

  const calcularResumo = () => {
    const emEstoque = produtos.filter(p => p.emEstoque);
    const totalBruto = emEstoque.reduce((acc, p) => acc + p.preco, 0);
    const desconto = cupons[cupom.toUpperCase()] || 0;
    const totalComDesconto = totalBruto * (1 - desconto);

    const totalFinal = totalComDesconto + frete;

    setResumo({
      produtos: emEstoque,
      totalBruto,
      desconto: totalBruto - totalComDesconto,
      frete,
      totalFinal
    });
  };

  return (

      <h2>🛒 Carrinho de Compras</h2>

      <ul>

          <li key={p.id}>{p.nome} - {formatarPreco(p.preco)}</li>
        ))}
      </ul>

      <div style={{ marginTop: '1em' }}>
        <label>
          Cupom de Desconto:
          <input
            type="text"
            value={cupom}
            onChange={(e) => setCupom(e.target.value)}

          />

      </div>

          <li>Frete: {formatarPreco(resumo.frete)}</li>
          <li><strong>Total: {formatarPreco(resumo.totalFinal)}</strong></li>
        </ul>
      )}
    </div>
  );
}

export default Carrinho;
```

```jsx
import React, { useState, useEffect } from "react";
import jsPDF from "jspdf";

function App() {
  const [produtos, setProdutos] = useState([]);
  const [carrinho, setCarrinho] = useState(() => {
    const carrinhoSalvo = localStorage.getItem("carrinho");
    return carrinhoSalvo ? JSON.parse(carrinhoSalvo) : [];
  });

  // Buscar produtos da API
  useEffect(() => {
    fetch("http://localhost:3001/produtos")
      .then(res => res.json())
      .then(data => setProdutos(data))
      .catch(err => console.error(err));
  }, []);

  // Salvar carrinho no localStorage sempre que mudar
  useEffect(() => {
    localStorage.setItem("carrinho", JSON.stringify(carrinho));
  }, [carrinho]);

  // Adicionar produto ao carrinho
  function adicionarAoCarrinho(produto) {
    setCarrinho(prev => [...prev, produto]);
  }

  // Remover produto pelo id do carrinho (remove a primeira ocorrência)
  function removerDoCarrinho(id) {
    const index = carrinho.findIndex(p => p.id === id);
    if (index === -1) return;
    const novoCarrinho = [...carrinho];
    novoCarrinho.splice(index, 1);
    setCarrinho(novoCarrinho);
  }

  // Função para formatar preço R$ 123,45
  function formatarPreco(valor) {
    return valor.toLocaleString("pt-BR", {
      style: "currency",
      currency: "BRL",
    });
  }

  // Calcular resumo (subtotal, desconto, frete, total)
  const resumo = React.useMemo(() => {
    const totalBruto = carrinho.reduce((acc, p) => acc + p.preco, 0);
    const desconto = totalBruto > 500 ? totalBruto * 0.1 : 0;
    const frete = totalBruto > 0 ? 30 : 0;
    const totalFinal = totalBruto - desconto + frete;
    return {
      produtos: carrinho,
      totalBruto,
      desconto,
      frete,
      totalFinal,
    };
  }, [carrinho]);

  // Exportar resumo para PDF
  function exportarParaPDF(resumo) {
    const doc = new jsPDF();
    doc.setFontSize(18);
    doc.text("Resumo do Carrinho", 10, 15);

    resumo.produtos.forEach((produto, i) => {
      const y = 25 + i * 10;
      doc.setFontSize(12);
      doc.text(`${produto.nome} - ${formatarPreco(produto.preco)}`, 10, y);
    });

    const yBase = 30 + resumo.produtos.length * 10;
    doc.text(`Subtotal: ${formatarPreco(resumo.totalBruto)}`, 10, yBase);
    doc.text(`Desconto: ${formatarPreco(resumo.desconto)}`, 10, yBase + 10);
    doc.text(`Frete: ${formatarPreco(resumo.frete)}`, 10, yBase + 20);

    doc.setFontSize(16);
    doc.text(`Total Final: ${formatarPreco(resumo.totalFinal)}`, 10, yBase + 40);

    doc.save("resumo_carrinho.pdf");
  }

  // Simular pagamento e enviar pedido para backend
  function simularPagamento() {
    if (resumo.totalFinal === 0) {
      alert("Carrinho vazio, adicione produtos antes de finalizar.");
      return;
    }

    alert("Processando pagamento...");

    fetch("http://localhost:3001/pedido", {
      method: "POST",
      headers: { "Content-Type": "application/json" },
      body: JSON.stringify(resumo),
    })
      .then(res => res.json())
      .then(data => {
        alert(data.mensagem);
        setCarrinho([]);
        localStorage.removeItem("carrinho");
      })
      .catch(err => {
        alert("Erro ao processar pagamento.");
        console.error(err);
      });
  }

  return (
    <div style={{ maxWidth: 600, margin: "20px auto", fontFamily: "Arial, sans-serif" }}>
      <h1>Carrinho de Compras</h1>

      <section>
        <h2>Produtos disponíveis</h2>
        <ul>
          {produtos.map(prod => (
            <li key={prod.id} style={{ marginBottom: 6 }}>
              {prod.nome} - {formatarPreco(prod.preco)}{" "}
              {prod.emEstoque ? (
                <button onClick={() => adicionarAoCarrinho(prod)}>Adicionar</button>
              ) : (
                <span style={{ color: "red" }}>Fora de estoque</span>
              )}
            </li>
          ))}
        </ul>
      </section>

      <section>
        <h2>Carrinho ({carrinho.length} itens)</h2>
        {carrinho.length === 0 ? (
          <p>Carrinho vazio.</p>
        ) : (
          <ul>
            {carrinho.map((prod, i) => (
              <li key={i} style={{ marginBottom: 6 }}>
                {prod.nome} - {formatarPreco(prod.preco)}{" "}
                <button onClick={() => removerDoCarrinho(prod.id)}>Remover</button>
              </li>
            ))}
          </ul>
        )}

        <div style={{ marginTop: 20 }}>
          <p>Subtotal: {formatarPreco(resumo.totalBruto)}</p>
          <p>Desconto: {formatarPreco(resumo.desconto)}</p>
          <p>Frete: {formatarPreco(resumo.frete)}</p>
          <h3>Total: {formatarPreco(resumo.totalFinal)}</h3>
        </div>

        <button onClick={() => exportarParaPDF(resumo)} style={{ marginRight: 10 }}>
          Exportar para PDF
        </button>

}
```

```js
const jwt = require("jsonwebtoken");

function autenticarToken(req, res, next) {
  const authHeader = req.headers["authorization"];
  const token = authHeader && authHeader.split(" ")[1];
  if (!token) return res.status(401).json({ error: "Token não fornecido" });

  jwt.verify(token, process.env.JWT_SECRET, (err, user) => {
```

```js
const jwt = require('jsonwebtoken');
const secret = 'seusegredoaqui';

// Middleware para verificar token
function authMiddleware(req, res, next) {
  const authHeader = req.headers.authorization;
  if (!authHeader) return res.status(401).json({ error: "Token não fornecido" });

  const token = authHeader.split(' ')[1];
  if (!token) return res.status(401).json({ error: "Token inválido" });

  jwt.verify(token, secret, (err, decoded) => {
    if (err) return res.status(401).json({ error: "Token inválido" });

    req.userId = decoded.id;
    req.userEmail = decoded.email;
    next();
  });
}
```

## 1. Criar arquivo `src/context/AuthContext.js`

```jsx
import React, { createContext, useState, useEffect } from "react";

// Cria o contexto
export const AuthContext = createContext();

// Provedor do contexto
export function AuthProvider({ children }) {
  const [token, setToken] = useState(() => localStorage.getItem("token"));
  const [user, setUser] = useState(null);
  const [loading, setLoading] = useState(true);

  // Função para carregar dados do usuário a partir do token
  const fetchUser = async (token) => {
    try {
      const res = await fetch("http://localhost:5000/api/auth/profile", {
        headers: { Authorization: `Bearer ${token}` },
      });

      if (!res.ok) throw new Error("Token inválido");

      const data = await res.json();
      setUser(data);
    } catch {
      logout();
    }
  };

  // Efeito para verificar token e carregar dados do usuário no carregamento
  useEffect(() => {
    if (token) {
      fetchUser(token).finally(() => setLoading(false));
    } else {
      setLoading(false);
    }
  }, [token]);

  // Função para login (armazenar token e buscar dados)
  const login = (newToken) => {
    localStorage.setItem("token", newToken);
    setToken(newToken);
  };

  // Função para logout (limpar token e dados)
  const logout = () => {
    localStorage.removeItem("token");
    setToken(null);
    setUser(null);
  };

  return (
    <AuthContext.Provider
      value={{ token, user, login, logout, loading }}
    >
      {children}
    </AuthContext.Provider>
  );
}
```

# 1. Implementando Refresh Token no React com Contexto

---

### Por que usar Refresh Token?

* Tokens JWT geralmente têm tempo de vida curto por segurança.
* Quando o token expira, o usuário é desconectado — ruim para UX.
* O Refresh Token permite solicitar um novo token sem pedir login novamente.

---

### Estratégia básica:

* No backend: gerar e enviar **Access Token** e **Refresh Token**.
* No frontend: armazenar Access Token no contexto/memória (não localStorage para segurança) e Refresh Token em httpOnly cookie (para evitar acesso via JS).
* Quando o Access Token expira, usar o Refresh Token para obter um novo Access Token automaticamente.

---

### Ajustando o frontend

**Atenção:** para armazenamento seguro, o Refresh Token deve ficar em cookie httpOnly configurado pelo backend (não acessível via JS).

---

### 1.1. Modificar AuthContext para incluir refresh automático

```jsx
import React, { createContext, useState, useEffect } from "react";

export const AuthContext = createContext();

export function AuthProvider({ children }) {
  const [token, setToken] = useState(null);
  const [user, setUser] = useState(null);
  const [loading, setLoading] = useState(true);

  // Função para buscar novo token via refresh token
  const refreshAccessToken = async () => {
    try {
      const res = await fetch("http://localhost:5000/api/auth/refresh_token", {
        method: "POST",
        credentials: "include", // envia cookies
      });

      if (!res.ok) throw new Error("Não foi possível renovar token");

      const data = await res.json();
      setToken(data.accessToken);
      return data.accessToken;
    } catch (error) {
      logout();
      return null;
    }
  };

  // Função para carregar dados do usuário
  const fetchUser = async (accessToken) => {
    try {
      const res = await fetch("http://localhost:5000/api/auth/profile", {
        headers: { Authorization: `Bearer ${accessToken}` },
      });

      if (!res.ok) throw new Error("Token inválido");

      const data = await res.json();
      setUser(data);
    } catch {
      logout();
    }
  };

  // Efeito inicial que tenta usar refresh token para obter acesso
  useEffect(() => {
    async function init() {
      const newToken = await refreshAccessToken();
      if (newToken) {
        await fetchUser(newToken);
      }
      setLoading(false);
    }
    init();
  }, []);

  // Função login atualizada para salvar token no estado (não localStorage)
  const login = (accessToken) => {
    setToken(accessToken);
  };

  // Logout limpa tudo
  const logout = () => {
    setToken(null);
    setUser(null);
    // backend deve limpar cookie refresh token
    fetch("http://localhost:5000/api/auth/logout", {
      method: "POST",
      credentials: "include",
    });
  };

  return (
    <AuthContext.Provider
      value={{ token, user, login, logout, loading, refreshAccessToken }}
    >
      {children}
    </AuthContext.Provider>
  );
}
```

### 1.2. Interceptar chamadas API para renovar token automaticamente

Você pode criar um helper para chamadas API que:

* Usa o token atual.
* Se receber 401 (token expirado), tenta refresh e reexecuta a chamada.

```jsx
export async function fetchWithAuth(url, options = {}, refreshAccessToken) {
  let token = options.token;

  if (!token) throw new Error("Token é necessário");

  const fetchOptions = {
    ...options,
    headers: {
      ...(options.headers || {}),
      Authorization: `Bearer ${token}`,
    },
  };

  let res = await fetch(url, fetchOptions);

  if (res.status === 401) {
    // tenta renovar token
    const newToken = await refreshAccessToken();
    if (!newToken) throw new Error("Não autorizado");

    // refaz a chamada com novo token
    const retryOptions = {
      ...options,
      headers: {
        ...(options.headers || {}),
        Authorization: `Bearer ${newToken}`,
      },
    };
    res = await fetch(url, retryOptions);
  }

  return res;
}
```

No componente você usaria assim:

```jsx
const { token, refreshAccessToken } = useContext(AuthContext);

const res = await fetchWithAuth("http://localhost:5000/api/some_protected_route", { method: "GET", token }, refreshAccessToken);
```

# 2. Armazenamento Seguro do Token

* Nunca salve Access Token em localStorage ou sessionStorage (vulnerável a XSS).
* Prefira guardar Access Token na memória (estado React).
* Refresh Token deve ser httpOnly cookie, gerenciado pelo backend, para não ser acessível via JS.
* Isso melhora a segurança e ajuda a proteger contra ataques.

```js
// server.js
require("dotenv").config();
const express = require("express");
const jwt = require("jsonwebtoken");
const bcrypt = require("bcryptjs");
const cookieParser = require("cookie-parser");
const cors = require("cors");

const app = express();
app.use(express.json());
app.use(cookieParser());

// Ajuste conforme seu frontend
app.use(
  cors({
    origin: "http://localhost:3000",
    credentials: true,
  })
);

// Mock de usuários (em DB real, use banco de dados)
const users = [
  {
    id: 1,
    email: "user@example.com",
    passwordHash: bcrypt.hashSync("senha123", 8), // senha hash
  },
];

// Armazena Refresh Tokens (em produção, use DB)
let refreshTokens = [];

// Funções para gerar tokens
function generateAccessToken(user) {
  return jwt.sign({ id: user.id, email: user.email }, process.env.ACCESS_TOKEN_SECRET, {
    expiresIn: "15m",
  });
}

function generateRefreshToken(user) {
  return jwt.sign({ id: user.id, email: user.email }, process.env.REFRESH_TOKEN_SECRET, {
    expiresIn: "7d",
  });
}

// Login
app.post("/api/auth/login", (req, res) => {
  const { email, password } = req.body;

  const user = users.find((u) => u.email === email);
  if (!user) return res.status(401).json({ error: "Credenciais inválidas" });

  const validPass = bcrypt.compareSync(password, user.passwordHash);
  if (!validPass) return res.status(401).json({ error: "Credenciais inválidas" });

  const accessToken = generateAccessToken(user);
  const refreshToken = generateRefreshToken(user);

  // Salva Refresh Token para controlar logout e revogação
  refreshTokens.push(refreshToken);

  // Envia refresh token em cookie httpOnly (não acessível via JS)
  res.cookie("refreshToken", refreshToken, {
    httpOnly: true,
    secure: false, // true em produção com https
    sameSite: "strict",
    maxAge: 7 * 24 * 60 * 60 * 1000, // 7 dias
  });

  return res.json({ accessToken });
});

// Endpoint para renovar token via Refresh Token
app.post("/api/auth/refresh_token", (req, res) => {
  const token = req.cookies.refreshToken;

  if (!token) return res.status(401).json({ error: "Sem token de refresh" });
  if (!refreshTokens.includes(token)) return res.status(403).json({ error: "Refresh token inválido" });

  try {
    const userData = jwt.verify(token, process.env.REFRESH_TOKEN_SECRET);
    const user = users.find((u) => u.id === userData.id);
    if (!user) return res.status(403).json({ error: "Usuário não encontrado" });

    const accessToken = generateAccessToken(user);
    return res.json({ accessToken });
  } catch (err) {
    return res.status(403).json({ error: "Refresh token expirado" });
  }
});

// Logout - remove refresh token
app.post("/api/auth/logout", (req, res) => {
  const token = req.cookies.refreshToken;
  refreshTokens = refreshTokens.filter((t) => t !== token);

  res.clearCookie("refreshToken");
  return res.json({ message: "Deslogado com sucesso" });
});

// Middleware para verificar Access Token
function authenticateToken(req, res, next) {
  const authHeader = req.headers["authorization"];
  const token = authHeader && authHeader.split(" ")[1];
  if (!token) return res.status(401).json({ error: "Token necessário" });

  jwt.verify(token, process.env.ACCESS_TOKEN_SECRET, (err, user) => {
    if (err) return res.status(403).json({ error: "Token inválido" });
    req.user = user;
    next();
  });
}

// Rota protegida exemplo
app.get("/api/auth/profile", authenticateToken, (req, res) => {
  const user = users.find((u) => u.id === req.user.id);
  if (!user) return res.status(404).json({ error: "Usuário não encontrado" });

  return res.json({ id: user.id, email: user.email });
});

app.listen(5000, () => console.log("Servidor rodando na porta 5000"));
```

```jsx
// src/AuthContext.js
import React, { createContext, useState, useEffect } from "react";

export const AuthContext = createContext();

export function AuthProvider({ children }) {
  const [accessToken, setAccessToken] = useState(null);
  const [loading, setLoading] = useState(true);

  // Função para renovar access token usando refresh token cookie
  async function refreshToken() {
    try {
      const res = await fetch("http://localhost:5000/api/auth/refresh_token", {
        method: "POST",
        credentials: "include", // envia cookies
      });
      if (!res.ok) throw new Error("Falha ao renovar token");
      const data = await res.json();
      setAccessToken(data.accessToken);
      return true;
    } catch (err) {
      setAccessToken(null);
      return false;
    }
  }

  // Tenta renovar token ao iniciar app
  useEffect(() => {
    refreshToken().finally(() => setLoading(false));
  }, []);

  // Login
  async function login(email, password) {
    const res = await fetch("http://localhost:5000/api/auth/login", {
      method: "POST",
      headers: { "Content-Type": "application/json" },
      credentials: "include", // para receber o cookie httpOnly
      body: JSON.stringify({ email, password }),
    });
    if (!res.ok) throw new Error("Credenciais inválidas");

    const data = await res.json();
    setAccessToken(data.accessToken);
  }

  // Logout
  async function logout() {
    await fetch("http://localhost:5000/api/auth/logout", {
      method: "POST",
      credentials: "include",
    });
    setAccessToken(null);
  }

  return (
    <AuthContext.Provider
      value={{ accessToken, login, logout, refreshToken, loading }}
    >
      {children}
    </AuthContext.Provider>
  );
}
```

## 2. Criar arquivo `axiosInstance.js` para configurar o cliente Axios com interceptors

```js
// src/axiosInstance.js
import axios from "axios";
import { getAccessToken, setAccessToken, clearAccessToken } from "./tokenService";

const instance = axios.create({
  baseURL: "http://localhost:5000/api",
  withCredentials: true, // enviar cookies automaticamente
});

// Flag para controlar renovação de token única vez por chamada
let isRefreshing = false;
let failedQueue = [];

const processQueue = (error, token = null) => {
  failedQueue.forEach(prom => {
    if (error) {
      prom.reject(error);
    } else {
      prom.resolve(token);
    }
  });

  failedQueue = [];
};

// Interceptor para adicionar token em todas as requisições
instance.interceptors.request.use(
  (config) => {
    const token = getAccessToken();
    if (token) {
      config.headers["Authorization"] = `Bearer ${token}`;
    }
    return config;
  },
  (error) => Promise.reject(error)
);

// Interceptor para lidar com respostas 401 (token expirado)
instance.interceptors.response.use(
  (response) => response,
  (error) => {
    const originalRequest = error.config;

    if (
      error.response &&
      error.response.status === 401 &&
      !originalRequest._retry
    ) {
      if (isRefreshing) {
        // Se já está renovando, aguarda o resultado
        return new Promise(function (resolve, reject) {
          failedQueue.push({ resolve, reject });
        })
          .then((token) => {
            originalRequest.headers["Authorization"] = "Bearer " + token;
            return instance(originalRequest);
          })
          .catch((err) => Promise.reject(err));
      }

      originalRequest._retry = true;
      isRefreshing = true;

      // Faz a requisição para renovar o token usando refresh_token cookie
      return new Promise(function (resolve, reject) {
        axios
          .post("http://localhost:5000/api/auth/refresh_token", {}, { withCredentials: true })
          .then(({ data }) => {
            setAccessToken(data.accessToken);
            instance.defaults.headers.common["Authorization"] =
              "Bearer " + data.accessToken;
            originalRequest.headers["Authorization"] = "Bearer " + data.accessToken;
            processQueue(null, data.accessToken);
            resolve(instance(originalRequest));
          })
          .catch((err) => {
            processQueue(err, null);
            clearAccessToken();
            reject(err);
          })
          .finally(() => {
            isRefreshing = false;
          });
      });
    }

    return Promise.reject(error);
  }
);

export default instance;
```

## 4. Atualizar `AuthContext.js` para usar axiosInstance

```jsx
// src/AuthContext.js
import React, { createContext, useState, useEffect } from "react";
import axios from "./axiosInstance";
import { setAccessToken, clearAccessToken } from "./tokenService";

export const AuthContext = createContext();

export function AuthProvider({ children }) {
  const [loading, setLoading] = useState(true);
  const [user, setUser] = useState(null);

  async function refreshToken() {
    try {
      const res = await axios.post("/auth/refresh_token");
      setAccessToken(res.data.accessToken);
      await fetchProfile();
      return true;
    } catch {
      clearAccessToken();
      setUser(null);
      return false;
    }
  }

  async function login(email, password) {
    const res = await axios.post("/auth/login", { email, password });
    setAccessToken(res.data.accessToken);
    await fetchProfile();
  }

  async function logout() {
    await axios.post("/auth/logout");
    clearAccessToken();
    setUser(null);
  }

  async function fetchProfile() {
    try {
      const res = await axios.get("/auth/profile");
      setUser(res.data);
    } catch {
      setUser(null);
    }
  }

  useEffect(() => {
    refreshToken().finally(() => setLoading(false));
  }, []);

  return (
    <AuthContext.Provider value={{ user, loading, login, logout }}>
      {children}
    </AuthContext.Provider>
  );
}
```
